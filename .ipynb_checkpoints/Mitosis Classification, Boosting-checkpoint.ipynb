In our previous code, we first segmented nuclei in order to create a two-stage general purpose mitosis detector - first find nuclei, then detect whether or not the nuclei is mitotic. This approach worked reasonably well on the data (it seemed to have good AUC on an unseen test set) but still has an atrociously low FP rate. Perhaps by using a boosting technique to focus on normal samples that are particularly hard to classify, we can improve these accuracies further.

Now that we have these datasets, let's train a classifier to distinguish between the two sets. Because of the scope of the problem, we should apply some percolation to the data.

```python
from skimage.transform import rotate

def variations(img_array, img_data):
    arr = []
    data = []
    for img, val in zip(img_array, img_data):
        for i in range(0,12):
            data.append(val)
        arr.append(img)
        arr.append(rotate(img, 90))
        arr.append(rotate(img, 180))
        arr.append(rotate(img, 270))
        arr.append(np.fliplr(img))
        arr.append(rotate(np.fliplr(img), 90))
        arr.append(rotate(np.fliplr(img), 180))
        arr.append(rotate(np.fliplr(img), 270))
        arr.append(np.flipud(img))
        arr.append(rotate(np.flipud(img), 90))
        arr.append(rotate(np.flipud(img), 180))
        arr.append(rotate(np.flipud(img), 270))
    return np.array(arr), np.array(data)
```

```python
from sknn.mlp import Classifier, Layer, Convolution
from sklearn import cross_validation

nn_old = nn

trainImg = trainImg.astype(np.float32)
trainInt = trainVal.astype(np.uint8)

print trainImg.shape

print "Building Classifier:"
nn = Classifier(
    layers=[
        Convolution("Rectifier", channels=128, kernel_shape=(4,4), pool_shape = (2, 2)),
        Convolution("Rectifier", channels=192, kernel_shape=(4,4), pool_shape = (2, 2), dropout = 0.3),
        Convolution("Rectifier", channels=256, kernel_shape=(4,4), pool_shape = (2, 2), dropout = 0.3),
        Layer("Rectifier", units=312, dropout = 0.4),
        Layer("Softmax")],
    #weights=nn_old.weights,
    learning_rate=0.001, verbose = True,
    n_iter=30)
```

```python
print "Training Classifier: 70/30 split"
X_train, X_test, y_train, y_test = cross_validation.train_test_split(trainImg[fullSet], trainInt[fullSet])

X_train, y_train = variations(X_train, y_train)
X_test, y_test = variations(X_test, y_test)

print X_train.shape
print y_train.shape
```

```python
nn.fit(X_train, y_train)
score = nn.score(X_test, y_test)
print score
```

```python
testImg = np.load("testImg_stage2.npy")
testVal = np.load("testVal_stage2.npy")
print testImg.shape
print sum(testVal)
```

```python
# Calculate AUC
from sklearn.metrics import roc_auc_score

probs = nn.predict_proba(testImg)
probs_filtered = []
for prob in probs:
    probs_filtered.append(prob[1])

auc = roc_auc_score(testVal, probs_filtered, average = 'macro')
print "AUC",auc
```

```python
from sklearn.metrics import f1_score

thresholds = np.divide(np.array(range(0, 1000)), 1000)
vals = []
for thresh in thresholds:
    classes = []
    for prob in probs_filtered:
        if prob > thresh:
            classes.append(1);
        else:
            classes.append(0);

    f1score = f1_score(testVal, np.array(classes), average='binary')
    vals.append(f1score)
plt.scatter(thresholds, vals)
```
